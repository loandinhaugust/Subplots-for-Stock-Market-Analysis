```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

start_date = "2018-12-31"
end_date = "2023-01-16"
cbaax = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/CBA.AX?period1=0&period2=9999999999&interval=1d&events=history&crumb=v41mcfpBc6K",parse_dates=["Date"], index_col="Date", usecols=["Date", "Open", "High", "Low", "Close", "Volume"], na_values=["nan"])
cbaax = cbaax.loc[start_date:end_date]

wbcax = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/WBC.AX?period1=0&period2=9999999999&interval=1d&events=history&crumb=v41mcfpBc6K",parse_dates=["Date"], index_col="Date", usecols=["Date", "Open", "High", "Low", "Close", "Volume"], na_values=["nan"])
wbcax = wbcax.loc[start_date:end_date]

anzax = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/ANZ.AX?period1=0&period2=9999999999&interval=1d&events=history&crumb=v41mcfpBc6K",parse_dates=["Date"], index_col="Date", usecols=["Date", "Open", "High", "Low", "Close", "Volume"], na_values=["nan"])
anzax = anzax.loc[start_date:end_date]

wesax = pd.read_csv("https://query1.finance.yahoo.com/v7/finance/download/WES.AX?period1=0&period2=9999999999&interval=1d&events=history&crumb=v41mcfpBc6K",parse_dates=["Date"], index_col="Date", usecols=["Date", "Open", "High", "Low", "Close", "Volume"], na_values=["nan"])
wesax = wesax.loc[start_date:end_date]


```

```python
#Historical view of closing price

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(20,10))

#Title of the figure
fig.suptitle('A historical view of closing price', fontsize=36)


# Plot the data in the first subplot
axs[0, 0].plot(cbaax['Close'],linewidth=0.75)
axs[0, 0].set_title('CommonWealth')


# Plot the data in the second subplot
axs[0, 1].plot(wbcax['Close'],color = 'purple',linewidth=0.75)
axs[0, 1].set_title('Westpac')

# Plot the data in the third subplot
axs[1, 0].plot(anzax['Close'],color = 'orange',linewidth=0.75)
axs[1, 0].set_title('ANZ')

# Plot the data in the fourth subplot
axs[1, 1].plot(wesax['Close'],color = 'green',linewidth=0.75)
axs[1, 1].set_title('Wesfarmers')

# Adjust the spacing between the subplots

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.85, wspace=0.2, hspace=0.25)

# Show the plot

plt.show()
```

```python
#Total volume of stock being traded each day

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(20,10))

#Title of the figure
fig.suptitle('Total volume of stock being traded each day', fontsize=36)


# Plot the data in the first subplot
axs[0, 0].plot(cbaax['Volume'],linewidth=0.75)
axs[0, 0].set_title('CommonWealth')


# Plot the data in the second subplot
axs[0, 1].plot(wbcax['Volume'],color = 'purple',linewidth=0.75)
axs[0, 1].set_title('Westpac')

# Plot the data in the third subplot
axs[1, 0].plot(anzax['Volume'],color = 'orange',linewidth=0.75)
axs[1, 0].set_title('ANZ')

# Plot the data in the fourth subplot
axs[1, 1].plot(wesax['Volume'],color = 'green',linewidth=0.75)
axs[1, 1].set_title('Wesfarmers')

# Adjust the spacing between the subplots

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.85, wspace=0.2, hspace=0.25)

# Show the plot

plt.show()
```
